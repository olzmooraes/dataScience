```python
import pandas as pd
df = pd.read_csv('csv/aluguel_uma_variavel_com_ruido.csv')
df.head()
```

```python

from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
treino, teste = train_test_split(df, test_size=0.3, random_state=42)

modelo = LinearRegression()
x_train = treino[['Area']]
y_train = treino[['Preco']]
x_test = teste[['Area']]
y_test = teste[['Preco']]
modelo = LinearRegression()
modelo.fit(x_train, y_train)
y_pred = modelo.predict(x_test)
print(f"R²: {r2_score(y_test, y_pred):.2f}")
```

```python
import matplotlib.pyplot as plt
plt.plot(x_test, y_pred, color='black')
plt.scatter(x_test, y_test)
plt.xlabel('Área (m²)')
plt.ylabel('Preço do Aluguel (R$)')
plt.title('Preço do Aluguel x Área da Casa')
plt.show()
```
