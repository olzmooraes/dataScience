```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
import pandas as pd
# Carregar o conjunto de dados Iris

iris = datasets.load_iris()
X = iris.data
y = iris.target

iris_pd = pd.DataFrame(iris.data, columns=iris.feature_names)

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# instanciando o função de normalizar as variaveis
scaler = StandardScaler()

# normalizando as variaveis de treino e teste
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)

# instanciando modelo de regressão logistica
modelo = LogisticRegression()
# treinando o modelo
modelo.fit(x_train_scaled, y_train)

# fazendo as previsões de Y(rotulo)
predict_scaled =  modelo.predict(x_test_scaled)
predict =  modelo.predict(x_test)

# vendo o erro médio
print("Accuracy normalizado:", accuracy_score(y_test, predict_scaled) )
print("Accuracy não normalizado:", accuracy_score(y_test, predict) )


```

```python
import matplotlib.pyplot as plt
plt.scatter(x_test_scaled[:, 0], x_test_scaled[:, 1],c=predict_scaled, cmap="viridis")
# plt.scatter(x_test, y_test)
plt.xlabel('plot dos dados normalizados')
# plt.ylabel('Preço do Aluguel (R$)')
# plt.title('Preço do Aluguel x Área da Casa') 
plt.show()
```
